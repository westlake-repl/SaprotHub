- This is **ColabProSST**, an interactive Google Colab tool powered by [ProSST](https://github.com/ai4protein/ProSST), a NeurIPS 2024 protein language model that combines amino-acid tokens and quantized structure tokens through sequence-structure disentangled attention. It supports the complete official quantized model family: [ProSST-20](https://huggingface.co/AI4Protein/ProSST-20), [ProSST-128](https://huggingface.co/AI4Protein/ProSST-128), [ProSST-512](https://huggingface.co/AI4Protein/ProSST-512), [ProSST-1024](https://huggingface.co/AI4Protein/ProSST-1024), [ProSST-2048](https://huggingface.co/AI4Protein/ProSST-2048), and [ProSST-4096](https://huggingface.co/AI4Protein/ProSST-4096). ProSST-2048 remains the default.

- ColabProSST integrates ProSST into the [SaprotHub](https://github.com/westlake-repl/SaprotHub) interactive workflow and the [ProSSTHub](https://huggingface.co/ProSSTHub) community ecosystem. It supports PDB/mmCIF structure quantization, zero-shot mutation-effect prediction, single-site saturation mutagenesis with heatmaps, protein-level classification and regression, residue-level classification, protein-pair classification and regression, checkpoint continuation, LoRA/PEFT training, direct ProSSTHub and compatible Hugging Face community model loading, protein-level and residue-level embedding extraction, independent prediction, CSV export, and model sharing through personal Hugging Face repositories with contribution to ProSSTHub.

- ColabProSST follows the accessible ColabPLM workflow promoted by [OPMC](https://theopmc.github.io/). Explore related tools: [ColabSaprot](https://colab.research.google.com/github/westlake-repl/SaprotHub/blob/main/colab/SaprotHub_v2.ipynb), [ColabSeprot](https://colab.research.google.com/github/westlake-repl/SaprotHub/blob/main/colab/ColabSeprot.ipynb), [ColabESMC](https://colab.research.google.com/github/westlake-repl/SaprotHub/blob/main/colab/ESM-Play-VC.ipynb), and [ColabProtT5](https://colab.research.google.com/github/westlake-repl/SaprotHub/blob/main/colab/ColabProtT5.ipynb).

### GPU selection guide

- **T4 GPU:** <font color="red">**Free to use**</font>. Suitable for structure-token conversion, zero-shot mutation prediction, protein property prediction, and small-batch training with the ProSST backbone frozen. <font color="red">**The free GPU can be unstable, so you may lose the connection or encounter memory limits during training and prediction. Use a small batch size, or switch to L4/A100 with Colab Pro.**</font>

- **L4 GPU (requires Colab Pro):** Provides more memory and is recommended for longer proteins, larger batches, or more demanding fine-tuning, but is slower than A100.

- **A100 GPU (requires Colab Pro):** The most powerful option. Recommended for full-backbone fine-tuning, large datasets, or memory-intensive experiments.


```python
#@title **Click the run-button**
#@markdown ### Hint:
#@markdown - It takes 3-8 minutes for the initial installation. The first installation restarts the Python kernel once to activate binary packages safely; after Colab reconnects, run this same cell once more.
#@markdown - The run-button's inactive state (<img src='https://github.com/westlake-repl/SaProtHub/blob/dev/Figure/run_button.png?raw=true' height='25px' width='25px' align='center'>) means the program is on standby. After you click it, the button shows a loading animation (<img src='https://github.com/westlake-repl/SaProtHub/blob/dev/Figure/run_button_working.png?raw=true' height='25px' width='25px' align='center'>) while ColabProSST starts.
#@markdown ### <font color=red>Important:</font>
#@markdown - Keep this cell running while using the interface. Stop the cell before changing the runtime type or disconnecting the runtime.
#@markdown - For unexpected connection or environment issues, clear the old runtime from **Runtime > Manage sessions**, reconnect, and run this cell again.
import importlib
import os
import signal
import shutil
import subprocess
import sys
import warnings
from pathlib import Path

SAPROTHUB_REPO = 'https://github.com/westlake-repl/SaprotHub.git'
SAPROTHUB_BRANCH = 'main'
PROSST_REPO = 'https://github.com/ai4protein/ProSST.git'

# Colab notebooks can be launched while the process is inside an old checkout.
# Always anchor managed files at /content, matching ColabSaprot's layout.
ROOT = Path('/content') if Path('/content').is_dir() else Path.cwd().resolve()
ROOT.mkdir(parents=True, exist_ok=True)
os.chdir(ROOT)
SAPROT_HOME = ROOT / 'SaprotHub'
PROSST_HOME = ROOT / 'ProSST'
PYG_HOME = ROOT / '.cache' / 'pyg'
HF_HOME = ROOT / '.cache' / 'huggingface'
ENVIRONMENT_GENERATION = '2026-07-12-clean-kernel-v2'
ENVIRONMENT_MARKER = ROOT / '.cache' / 'colabprosst' / 'environment_generation'
SAPROT_REQUIRED = [
    Path('pyproject.toml'),
    Path('saprot/__init__.py'),
    Path('saprot/utils/colab_prosst_workflow.py'),
    Path('saprot/utils/colab_prosst_ui.py'),
    Path('saprot/scripts/plot_prosst_saturation.py'),
]
PROSST_REQUIRED = [
    Path('prosst/structure/get_sst_seq.py'),
    Path('prosst/structure/static/AE.pt'),
    Path('prosst/structure/static/20.joblib'),
    Path('prosst/structure/static/128.joblib'),
    Path('prosst/structure/static/512.joblib'),
    Path('prosst/structure/static/1024.joblib'),
    Path('prosst/structure/static/2048.joblib'),
    Path('prosst/structure/static/4096.joblib'),
]

for path in [PYG_HOME, HF_HOME, ENVIRONMENT_MARKER.parent, ROOT / 'prosst_uploads', ROOT / 'prosst_structure_assets']:
    path.mkdir(parents=True, exist_ok=True)
os.environ['PYG_HOME'] = str(PYG_HOME)
os.environ['HF_HOME'] = str(HF_HOME)
os.environ.pop('TRANSFORMERS_CACHE', None)
os.environ['PYTHONWARNINGS'] = 'ignore::FutureWarning'
os.environ['TOKENIZERS_PARALLELISM'] = 'false'
os.environ['MPLBACKEND'] = 'Agg'
os.environ['PROSST_HOME'] = str(PROSST_HOME)
warnings.filterwarnings('ignore', category=FutureWarning)

def checkout_complete(home, required_files):
    return home.is_dir() and all((home / relative).is_file() for relative in required_files)

def add_project_paths():
    for path in [SAPROT_HOME, PROSST_HOME]:
        value = str(path)
        if value not in sys.path:
            sys.path.insert(0, value)

def runtime_environment():
    env = os.environ.copy()
    project_paths = [str(SAPROT_HOME), str(PROSST_HOME)]
    if env.get('PYTHONPATH'):
        project_paths.append(env['PYTHONPATH'])
    env['PYTHONPATH'] = os.pathsep.join(project_paths)
    return env

def probe_runtime():
    if not checkout_complete(SAPROT_HOME, SAPROT_REQUIRED):
        return False, 'SaprotHub checkout is missing ColabProSST files.'
    if not checkout_complete(PROSST_HOME, PROSST_REQUIRED):
        return False, 'Official ProSST checkout is incomplete.'

    probe = subprocess.run(
        [
            sys.executable,
            '-c',
            (
                'import importlib; import saprot; import numpy; import pandas; '
                'import torch; import torch_geometric; import torch_scatter; '
                "import matplotlib; matplotlib.use('Agg'); import matplotlib.pyplot as plt; "
                'figure = plt.figure(); plt.close(figure); '
                'import ipywidgets; import jupyter_ui_poll; '
                "importlib.import_module('prosst.structure.get_sst_seq'); "
                'from saprot.utils.colab_prosst_workflow import ColabProSSTWorkflow; '
                'from saprot.utils.colab_prosst_ui import ColabProSSTUI'
            ),
        ],
        env=runtime_environment(),
        capture_output=True,
        text=True,
    )
    details = (probe.stderr or probe.stdout).strip()
    return probe.returncode == 0, details

def run_command(args):
    subprocess.run([str(arg) for arg in args], check=True)

def run_pip(*args):
    run_command([sys.executable, '-m', 'pip', *args])

def clone_saprothub():
    staging_home = ROOT / '.SaprotHub-installing'
    if staging_home.exists():
        shutil.rmtree(staging_home)
    try:
        run_command([
            'git', 'clone', '--depth', '1', '--branch', SAPROTHUB_BRANCH,
            SAPROTHUB_REPO, staging_home,
        ])
        if not checkout_complete(staging_home, SAPROT_REQUIRED):
            raise RuntimeError(
                f'{SAPROTHUB_REPO}@{SAPROTHUB_BRANCH} does not contain '
                'ColabProSST.'
            )
        if SAPROT_HOME.exists():
            shutil.rmtree(SAPROT_HOME)
        staging_home.replace(SAPROT_HOME)
    except Exception:
        if staging_home.exists():
            shutil.rmtree(staging_home)
        raise

def update_saprothub():
    if not checkout_complete(SAPROT_HOME, SAPROT_REQUIRED):
        return
    if not (SAPROT_HOME / '.git').is_dir():
        print('Reusing the existing ColabProSST source checkout.')
        return

    try:
        subprocess.run(
            [
                'git', '-C', str(SAPROT_HOME), 'fetch', '--depth', '50',
                'origin', SAPROTHUB_BRANCH,
            ],
            check=True,
            stdout=subprocess.DEVNULL,
            stderr=subprocess.DEVNULL,
        )
        current = subprocess.check_output(
            ['git', '-C', str(SAPROT_HOME), 'rev-parse', 'HEAD'],
            text=True,
        ).strip()
        latest = subprocess.check_output(
            ['git', '-C', str(SAPROT_HOME), 'rev-parse', 'FETCH_HEAD'],
            text=True,
        ).strip()
        if current != latest:
            print('Updating ColabProSST source...', flush=True)
            run_command([
                'git', '-C', SAPROT_HOME, 'merge', '--ff-only', 'FETCH_HEAD',
            ])
    except (subprocess.CalledProcessError, OSError) as exc:
        print(
            'Could not check for a ColabProSST source update; reusing the '
            f'existing checkout ({type(exc).__name__}).'
        )

def project_revision(home=SAPROT_HOME):
    try:
        return subprocess.check_output(
            ['git', '-C', str(home), 'rev-parse', '--short', 'HEAD'],
            text=True,
        ).strip()
    except (subprocess.CalledProcessError, OSError):
        return 'unknown'

def ensure_official_prosst():
    if checkout_complete(PROSST_HOME, PROSST_REQUIRED):
        return
    staging_home = ROOT / '.ProSST-installing'
    if staging_home.exists():
        shutil.rmtree(staging_home)
    try:
        run_command([
            'git', 'clone', '--depth', '1', PROSST_REPO, staging_home,
        ])
        if not checkout_complete(staging_home, PROSST_REQUIRED):
            raise RuntimeError(
                'The official ProSST checkout is incomplete after cloning.'
            )
        if PROSST_HOME.exists():
            shutil.rmtree(PROSST_HOME)
        staging_home.replace(PROSST_HOME)
    except Exception:
        if staging_home.exists():
            shutil.rmtree(staging_home)
        raise

if checkout_complete(SAPROT_HOME, SAPROT_REQUIRED):
    update_saprothub()

runtime_ready, probe_details = probe_runtime()
if not runtime_ready:
    print('Installing ProSST...', flush=True)
    if not checkout_complete(SAPROT_HOME, SAPROT_REQUIRED):
        clone_saprothub()
    ensure_official_prosst()

    # Match ColabSaprot's environment order, with only ProSST quantizer extras added.
    run_pip('install', '-r', SAPROT_HOME / 'requirements.txt')
    run_pip('install', SAPROT_HOME)
    run_pip('install', 'jupyter_ui_poll')
    run_pip('uninstall', '-y', 'torchao')
    run_pip('install', 'joblib==1.3.2', 'pathos==0.3.2', 'biotite==0.39.0')

    torch_info = subprocess.check_output(
        [
            sys.executable, '-c',
            "import torch; print(torch.__version__.split('+')[0]); print(torch.version.cuda or 'cpu')",
        ],
        text=True,
    ).strip().splitlines()
    torch_version, torch_cuda = torch_info
    if torch_version.startswith('2.2.'):
        torch_version = '2.2.0'
    pyg_backend = 'cpu' if torch_cuda == 'cpu' else 'cu' + torch_cuda.replace('.', '')
    pyg_wheel_url = f'https://data.pyg.org/whl/torch-{torch_version}+{pyg_backend}.html'
    print(f'Installing ProSST PyG extensions from {pyg_wheel_url}')
    run_pip(
        'install', '--no-index', 'pyg_lib', 'torch_scatter', 'torch_sparse',
        'torch_cluster', 'torch_spline_conv', '-f', pyg_wheel_url,
    )
    run_pip('install', 'torch-geometric==2.5.0')

    for path in [
        SAPROT_HOME / 'LMDB',
        SAPROT_HOME / 'output',
        SAPROT_HOME / 'datasets',
        SAPROT_HOME / 'adapters' / 'classification' / 'Local',
        SAPROT_HOME / 'adapters' / 'regression' / 'Local',
        SAPROT_HOME / 'structures',
    ]:
        path.mkdir(parents=True, exist_ok=True)
    for binary in (SAPROT_HOME / 'bin').glob('*'):
        binary.chmod(binary.stat().st_mode | 0o111)

    runtime_ready, probe_details = probe_runtime()
    if not runtime_ready:
        raise RuntimeError(
            'ColabProSST installation finished, but the isolated import check failed:\n'
            + probe_details
        )

environment_generation = (
    ENVIRONMENT_MARKER.read_text(encoding='utf-8').strip()
    if ENVIRONMENT_MARKER.is_file()
    else ''
)
if environment_generation != ENVIRONMENT_GENERATION:
    ENVIRONMENT_MARKER.write_text(ENVIRONMENT_GENERATION, encoding='utf-8')
    print('ColabProSST environment is installed. Restarting the Python kernel once', flush=True)
    print('After Colab reconnects, run this same cell once more.', flush=True)
    sys.stdout.flush()
    os.kill(os.getpid(), signal.SIGKILL)

# A rerun in the same Colab kernel must not keep modules from an old checkout.
for module_name in list(sys.modules):
    if (
        module_name == 'saprot'
        or module_name.startswith('saprot.')
        or module_name == 'prosst'
        or module_name.startswith('prosst.')
    ):
        del sys.modules[module_name]

add_project_paths()
importlib.invalidate_caches()
try:
    importlib.import_module('prosst.structure.get_sst_seq')
    from saprot.utils.colab_prosst_workflow import ColabProSSTWorkflow
    from saprot.utils.colab_prosst_ui import ColabProSSTUI
except ModuleNotFoundError as exc:
    raise RuntimeError(
        'ColabProSST source import failed after installation. '
        f'Runtime root: {ROOT}; SaprotHub: {SAPROT_HOME}; ProSST: {PROSST_HOME}. '
        'Clear the old runtime with Runtime > Manage sessions, reconnect, '
        'and run this cell again.'
    ) from exc

COLABPROSST_WORKFLOW = ColabProSSTWorkflow()
print(
    f'ColabProSST source: {SAPROTHUB_REPO}@{SAPROTHUB_BRANCH} '
    f'({project_revision()})'
)
print(f'Official ProSST source: {PROSST_REPO} ({project_revision(PROSST_HOME)})')
print('ProSST is installed successfully!')

COLABPROSST_UI = ColabProSSTUI(COLABPROSST_WORKFLOW)
COLABPROSST_UI.launch()
```
